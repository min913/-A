```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
#1.서울버스노선이용자수, 서울버스정류소위치 파일 가져와서 각각 df1, df2에 저장합니다. 만약 한글이 깨질 경우 구글링해서 문제를 해결해보세요!
df1= pd.read_csv('/content/drive/MyDrive/서울버스노선이용자수.csv')
df2= pd.read_csv('/content/drive/MyDrive/서울버스정류소위치.csv')
```

```python
#2. head()와 tail()을 통해 df1, df2 대략적으로 확인합니다, print말고 display 함수를 써보세요!
display(df1.head())
display(df1.tail())
display(df2.head())
display(df2.tail())
```

```python
#3.각 데이터프레임의 정보를 확인합니다. (서울버스노선이용자수 데이터(df1)에 결측치가 있는 것을 확인할 수 있습니다)
print(df1.info())
print(df2.info())
```

```python
#4. df1의 노선명 column 중 결측치가 있는 행만 확인해봅니다.(index가 427603인 행부터 결함이 있다는 것을 확인할 수 있습니다.)
df1[df1["노선명"].isnull()]
```

```python
# 개수가 2784개인 것을 보아 노선명에 결측치가 있는 행이 index가 427603~430386인 행이라고 추측할 수 있습니다.
print(df1.isnull().sum())
print('-'*50)
print(430386-427602)
```

```python
#5. df1에서 index가 427603인 행부터 삭제해서 새로운 df1을 만듭니다.

df1=df1.drop(index=range(427603, len(df1)))
```

```python
#아직 하차 정류장ARS와 하차_정류장 순번에 결측치가 있는 것을 확인할 수prin
df1.info()
```

```python
#6. 이번에는 df1에서 결측치가 있는 행을 모두 제거해 새로운 df1을 만듭니다.

df1=df1.dropna()
```

```python
# 결측치가 모두 없어진 것을 확인할 수 있습니다
df1.info()
```

```python
#7. df1와 df2를 merge하기 위해 ARS의 data type을 통일할 것입니다. df1의 승차_정류장 ARS와 하차_정류장 ARS의 dtype을 float에서 int로 바꿉니다. (람다식 사용!)
df1['승차_정류장ARS'] = df1['승차_정류장ARS'].apply(lambda x: int(x))
df1['하차_정류장ARS'] = df1['하차_정류장ARS'].apply(lambda x: int(x))
```

```python
#8. df2에서 ARS_ID, X좌표, Y좌표 column만 선택해 df2_sample을 만듭니다.
df2_sample = df2[['ARS_ID', 'X좌표', 'Y좌표']]
display(df2_sample )
```

```python
#9. ARS 번호를 이용하여 df1에 df2_sample을 merge해 '승차_정류장_X좌표', '승차_정류장_Y좌표', '하차_정류장_X좌표', '하차_정류장_Y좌표' 4개의 column을 추가하여 새로운 df1을 만듭니다
# merge할 때 inner 조인 방식을 사용합니다!

# '승차_정류장ARS'와 '하차_정류장ARS' 열을 기준으로 'df1'과 'df2_sample'을 inner join하여 merge 수행
merged_df = pd.merge(df1, df2_sample, how='inner', left_on='승차_정류장ARS', right_on='ARS_ID')
merged_df = pd.merge(merged_df, df2_sample, how='inner', left_on='하차_정류장ARS', right_on='ARS_ID', suffixes=('_승차', '_하차'))

# '승차_정류장_X좌표', '승차_정류장_Y좌표', '하차_정류장_X좌표', '하차_정류장_Y좌표' 열 추가
df1 = merged_df[['기준일자', '노선명', '승차_정류장ARS', '승차_정류장명', '하차_정류장ARS', '하차_정류장명', '승차_정류장순번', '하차_정류장순번', '승객수',
                 'X좌표_승차', 'Y좌표_승차', 'X좌표_하차', 'Y좌표_하차']]

# 새로운 열 이름으로 열 이름 변경
df1 = df1.rename(columns={'X좌표_승차': '승차_정류장_X좌표', 'Y좌표_승차': '승차_정류장_Y좌표',
                          'X좌표_하차': '하차_정류장_X좌표', 'Y좌표_하차': '하차_정류장_Y좌표'})

print(df1)

```

```python
#10. 승차정류장의 좌표와 하차정류장의 좌표를 이용해 거리를 구하고 df1에 새로운 column '거리'를 추가해줍니다!
#경도와 위도 1도의 거리는 133.33km라 가정하고 피타고라스의 정리를 사용해 직선거리를 구합니다.

import numpy as np

# 승차 정류장과 하차 정류장의 x 좌표 차이와 y 좌표 차이를 구함
x_diff = (df1['승차_정류장_X좌표'] - df1['하차_정류장_X좌표']) * 133.33
y_diff = (df1['승차_정류장_Y좌표'] - df1['하차_정류장_Y좌표']) * 133.33

# 거리 구하기 (피타고라스의 정리 사용)
df1['거리'] = np.sqrt(x_diff**2 + y_diff**2)
df1
```

```python
#11. 거리에 따라 내림차순으로 df1을 정렬하고 index를 초기화해줍니다. (제가 실수했는데.. index column도 없애주세요!)

df1 = df1.sort_values(by='거리', ascending=False)

df1 = df1.reset_index(drop=True)

display(df1)
```

```python
#12. 노선당 몇 개의 경로가 있는지 그리고 각각의 경로 거리의 min, max, mean 값을 나타내봅니다.
df1result = df1.groupby(by='노선명')

size = df1result.size()
min = df1result['거리'].min()
max = df1result['거리'].max()
mean = df1result['거리'].mean()
name=['size','min','mean','max']


resultframe=pd.DataFrame(zip(size,min,mean,max), index=df1['노선명'], columns=name)
display(resultframe)
```

```python
# 노선별로 그룹화하고 거리에 대한 최소, 최대, 평균 계산
df1root = df1.groupby('노선명').agg({'거리': ['size', 'min', 'mean', 'max']})

# 결과 출력
display(df1root)
```

```python
#13. 최종적으로 df1을 csv파일로 만들어줍니다!
#csv 파일에서 한글이 깨지지 않도록 구글링해서 적용해주세요
df1.to_csv('df1.csv', index=False, encoding = 'cp949')
```
